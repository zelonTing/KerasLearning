本文将使用keras已经训练好的模型，进行图像分类。

这些模型包括：
* VGG16
* VGG19
* ResNet50
* InceptionV3

```python
import tensorflow as tf
import keras

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from IPython.display import Image
```

## 1、VGG16
### 1.1 从网上获取一个要进行模型分类的图像
### 1.2 加载VGG模型 

```python
from keras.applications.vgg16 import VGG16
```

```python
model_vgg16 = VGG16() #如果本地没有模型文件，就会在网上下载一个
```

### 1.3加载，准备图像

```python

from keras.preprocessing.image import load_img
```

```python
image_file = load_img('elephant.jpg',target_size=(224,224))
```

```python
from keras.preprocessing.image import img_to_array
```

```python
image = img_to_array(image_file)
```

```python
image.shape
```

```python
image=image.reshape((1,image.shape[0],image.shape[1],image.shape[2]))
image.shape
```

```python
from keras.applications.vgg16 import preprocess_input
```

```python
image = preprocess_input(image)
```

### 1.4做一个预测

```python

y_pre = model_vgg16.predict(image)
```

```python
y_pre
```

### 1.5 解释预测结果

```python
from keras.applications.vgg16 import decode_predictions
```

```python
label = decode_predictions(y_pre)  #将机率转换为类别标签
```

```python
label
```
